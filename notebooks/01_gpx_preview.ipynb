```python
from pathlib import Path

GPX_DIR = Path("../data/raw/gpx")
gpx_files = sorted(GPX_DIR.glob("*.gpx"))

print("Найдено GPX файлов:", len(gpx_files))
for i, p in enumerate(gpx_files[:50], 1):
    print(f"{i:02d}. {p.name}")

# Выбор файла по номеру (поменяй index на нужный)
index = 1  # <-- поставь номер из списка
gpx_path = gpx_files[index-2]
print("\nВыбран файл:", gpx_path)
```

```python
import gpxpy
from datetime import datetime

with open(gpx_path, "r", encoding="utf-8") as f:
    gpx = gpxpy.parse(f)

tracks_cnt = len(gpx.tracks)
segs_cnt = sum(len(t.segments) for t in gpx.tracks)
points_cnt = sum(len(seg.points) for t in gpx.tracks for seg in t.segments)

# проверим наличие time и ele
has_time = False
has_ele = False
t_min = None
t_max = None

for t in gpx.tracks:
    for seg in t.segments:
        for pt in seg.points:
            if pt.time is not None:
                has_time = True
                if t_min is None or pt.time < t_min: t_min = pt.time
                if t_max is None or pt.time > t_max: t_max = pt.time
            if pt.elevation is not None:
                has_ele = True

print("Треков в файле:", tracks_cnt)
print("Сегментов всего:", segs_cnt)
print("Точек всего:", points_cnt)
print("Есть time:", has_time)
print("Есть ele:", has_ele)
print("Время старта:", t_min)
print("Время финиша:", t_max)
```
